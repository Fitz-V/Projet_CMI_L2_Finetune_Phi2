# Étape 3 : Évaluation sur DS-1000 et Analyse Critique

Nous évaluons les performances de notre modèle fine-tuné face à plusieurs baselines de l'industrie. Le score présenté est le Pass@1 (pourcentage de codes générés passant avec succès les tests unitaires d'exécution).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données issues de l'évaluation DS-1000
data = {
    "Modèle": ["GPT-4-turbo", "GPT-3.5-turbo (0125)", "GPT-3.5-turbo (0613)", 
               "Phi-2 Fine-tuné (Ours)", "Mistral 7B", "Phi-2 Base"],
    "Pass@1 (%)": [53.9, 39.4, 38.6, 10.2, 10.1, 10.0],
    "Catégorie": ["Propriétaire", "Propriétaire", "Propriétaire", 
                  "Open-Source (Fine-tuné)", "Open-Source (Base)", "Open-Source (Base)"]
}

df = pd.DataFrame(data)

# Génération du graphique
plt.figure(figsize=(10, 6))
sns.barplot(x="Pass@1 (%)", y="Modèle", hue="Catégorie", data=df, dodge=False, palette="viridis")
plt.title("Performances sur DS-1000 (Génération de code Data Science)")
plt.axvline(x=10.0, color='red', linestyle='--', alpha=0.5, label='Baseline Phi-2')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

## Analyse du "Distribution Shift" (Biais d'Alignement)

Malgré l'intégration de la syntaxe de `pandas` et `matplotlib`, le score global stagne autour de la baseline (10.2% vs 10.0%). 

**Pourquoi ?**
1. **Nature de l'entraînement :** Notre adaptateur LoRA a été entraîné en *Causal Language Modeling* pur sur du code source extrait de GitHub. Il se comporte comme un compléteur de code strict.
2. **Nature de l'évaluation :** Le benchmark DS-1000 soumet des instructions conversationnelles (ex: *"Write a function to..."*).

Le modèle échoue à franchir le cap des 10% non pas par manque de connaissances syntaxiques, mais parce qu'il n'a pas été aligné (*Instruction Tuning*) pour comprendre le format de la requête du benchmark.
